```python
# IMPORTING MODUELS

import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
# LOADING AND CONVERTING DATA

# importing data
data = pd.read_csv("car.data")
print(data.head())

#Converting no numeric data (Strings) into numbered data
le = preprocessing.LabelEncoder()

# Takes a list and returns to us an array containing new values
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))

# Recombing data in to feature list and a lable list using the Zip() function
X = list(zip(buying, maint, door, persons, lug_boot, safety))  # features
y = list(cls)

# Spliting data into training and testing data
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

```python
# TRAINING KNN CLASSIFIER

# Specifing how many neighbours to look for
model = KNeighborsClassifier(n_neighbors=9)

# training model
model.fit(x_train, y_train)

# Scoring the model
acc = model.score(x_test, y_test)
print(acc)
```

```python
# TESTING THE MODEL

# Predict using the model on the test data (x_test)
predicted = model.predict(x_test)

# Define the class names for better readability of predictions
names = ["unacc", "acc", "good", "vgood"]

# Loop through the predicted results and print them alongside actual data and labels
for x in range(len(predicted)):
    print("Predicted: ", names[predicted[x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
```

```python
# Print the predicted class label, data, and actual class label
# Find the 9 nearest neighbors of the current data point using the trained model
# Print the indices and distances of the nearest neighbors

for x in range(len(predicted)):
    print("Predicted: ", names[predicted[x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
    n = model.kneighbors([x_test[x]], 9, True)
    print("N: ", n)
```
